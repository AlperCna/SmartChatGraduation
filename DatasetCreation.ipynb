```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

```python
df = df[["text", "label"]]
df.head()
```

```python
df = df[df["label"].isin(["Positive", "Negative", "Notr"])]
df.head()
```

```python
df = df.dropna()
df = df[df["text"].str.len() > 5]
df.head()
```

```python
df_small = (
    df.groupby("label")
      .apply(lambda x: x.sample(1000, random_state=42))
      .reset_index(drop=True)
)

df_small["label"].value_counts()
```

```python
df_small.to_csv("smartchat_sentiment_3000.csv", index=False)
```
